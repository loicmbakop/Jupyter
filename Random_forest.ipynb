```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.model_selection import train_test_split

data =pd.read_csv('C:/Users/MBAKOP Loic/Documents/Machine Learning Python/des-fichiers-complementaires-4-335-ko/Chapitre_09/Data/spambase.csv',sep=',')
data
```

```python
email_class = data['1']

data.drop(['1'],axis = 1, inplace=True)

```

```python
x_train,x_test,y_train,y_test = train_test_split(data,email_class, test_size=0.25)
```

```python
alg_decisionTree = RandomForestClassifier(n_estimators=1, max_depth=60,n_jobs=-1)
dt_model = alg_decisionTree.fit(x_train,y_train)
```

```python
dt_predictions = dt_model.predict(x_test)
precision,recall,fscore,support =score(y_test, dt_predictions,pos_label=1,average='binary')
```

```python
print('Precision: {} / Recall: {} / Accuracy: {}'.format(round(precision, 3),
                                                         round(recall, 3), round((dt_predictions==y_test).sum() / len(dt_predictions),3)))
```

```python
from sklearn.model_selection import KFold, cross_val_score

alg_RandomForest = RandomForestClassifier(n_estimators=60,max_depth=60, n_jobs=-1)
k_fold = KFold(n_splits = 4)
cross_val_score(alg_RandomForest,x_test,y_test,cv=k_fold,scoring='recall',n_jobs=-1)
```
